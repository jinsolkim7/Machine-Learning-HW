### 1. 데이터 불러오기

```python
import numpy as np

x_train = np.loadtxt('./hw1_data/p2_train_input.txt')
y_train = np.loadtxt('./hw1_data/p2_train_target.txt')
x_test = np.loadtxt('./hw1_data/p2_test_input.txt')
y_test = np.loadtxt('./hw1_data/p2_test_target.txt')
```

```python
x_train.shape
```

### 2. 모델 만들기

- 모델 생성: Sequential()
- 모델을 생성한 직후에는 아무런 레이어가 추가되어 있지 않은 '빈 모델'
- 레이어 추가: add() 함수 이용

```python
from keras import models
from keras import layers

model = models.Sequential() # 비어 있는 모델
model.add(layers.Dense(32, activation = 'relu', input_shape=(2,))) # input layer
model.add(layers.Dense(32, activation = 'relu')) # hidden layer
model.add(layers.Dense(1, activation = 'sigmoid')) # output layer
```

```python
model.summary()
```

### 3. 훈련 검증 세트 준비

```python
x_val = x_train[:100]
partial_x_train = x_train[100:]
y_val = y_train[:100]
partial_y_train = y_train[100:]
```

### 4. 모델 컴파일

```python
from keras import optimizers

model.compile(optimizer = 'rmsprop',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])
```

### 5. 모델 학습

```python
history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs = 1000,
                    validation_data=(x_val, y_val))
```

```python
history_dict = history.history
```

```python
history_dict.keys()
```

### 6. 훈련-검증 손실/정확도 그래프 그리기

```python
# 훈련과 검증 손실 그리기
import matplotlib.pyplot as plt

history_dict = history.history
loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, 'b', label = 'Training loss')
plt.plot(epochs, val_loss, 'r', label = 'Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 훈련과 검증 정확도 그리기
import matplotlib.pyplot as plt

history_dict = history.history
acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']

plt.plot(epochs, acc, 'b', label = 'Training acc')
plt.plot(epochs, val_acc, 'r', label = 'Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.show()
```
