### 1. 데이터 불러오기

```python
import numpy as np

x_train = np.loadtxt('./hw1_data/p1_train_input.txt')
y_train = np.loadtxt('./hw1_data/p1_train_target.txt')
x_test = np.loadtxt('./hw1_data/p1_test_input.txt')
y_test = np.loadtxt('./hw1_data/p1_test_target.txt')
```

```python
x_train.shape
```

### 2. 모델 만들기

- 모델 생성: Sequential()
- 모델을 생성한 직후에는 아무런 레이어가 추가되어 있지 않은 '빈 모델'
- 레이어 추가: add() 함수 이용

```python
from keras import models
from keras import layers

model = models.Sequential() # 비어 있는 모델
model.add(layers.Dense(32, activation = 'relu', input_shape=(2,)))
model.add(layers.Dense(32, activation = 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

### 3. 훈련 검증 세트 준비

```python
x_val = x_train[:100]
partial_x_train = x_train[100:]
y_val = y_train[:100]
partial_y_train = y_train[100:]
```

### 4. 모델 컴파일

```python
from keras import optimizers

model.compile(optimizer = 'rmsprop',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])
```

### 5. 모델 학습

```python
history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs = 1000,
                    validation_data=(x_val, y_val))
```

```python
history_dict = history.history
```

```python
history_dict.keys()
```

### 6. 훈련-검증 손실/정확도 그래프 그리기

```python
# 훈련과 검증 손실 그리기
import matplotlib.pyplot as plt

history_dict = history.history
loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, 'b', label = 'Training loss')
plt.plot(epochs, val_loss, 'r', label = 'Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 훈련과 검증 정확도 그리기
import matplotlib.pyplot as plt

history_dict = history.history
acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']

plt.plot(epochs, acc, 'b', label = 'Training acc')
plt.plot(epochs, val_acc, 'r', label = 'Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.show()
```

(해석)
- 예상했듯이 training loss가 epoch마다 감소하고, training accuracy는 epoch마다 증가한다. 경사하강법 최적화를 사용했을 때 반복마다 최소화되는 것이 손실이므로 기대했던 대로이다.
- 하지만, validation loss와 validation accuracy는 이와 같지 않다. 약 600번째 epoch에서 그래프가 역전되는 현상을 보인다. 이것은 training dataset에서 잘 작동하는 모델이 처음 보는 데이터에서는 잘 작동하지 않는 경우 즉, 과대적합(overfitting)이 된 경우이다. 약 600번째 epoch 이후부터 training data에 과도하게 최적화되어 훈련 데이터에 특화된 표현을 학습하므로 훈련 세트 이외의 데이터에는 일반화되지 못한 것.
- 따라서, overfitting을 방지하기 위해서 600번째 epoch 이후에 훈련을 중지시킨다.
- 이제 처음부터 다시 새로운 신경망을 600번의 epoch 동안만 훈련하고 test data에서 평가해보자.

```python
model = models.Sequential()
model.add(layers.Dense(32, activation = 'relu', input_shape=(2,)))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model.fit(x_train, y_train, epochs=1000)
results = model.evaluate(x_test, y_test)

results
```
